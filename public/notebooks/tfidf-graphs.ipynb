```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 1. Set the visual style to match the 'whitegrid' look of the original image
sns.set_theme(style="whitegrid")
```

```python
# 2. Generate Data
# N is given as 100 in the title
N = 100

# Generate DF values from 1 to 99 (avoiding 0 to prevent division by zero errors)
# We use 500 points to make the curves smooth
df = np.linspace(1, 99, 500)

# Calculate the two IDF variations
# First series: Standard inverse proportion
idf_linear = N / df 
# Second series: Natural logarithm of the inverse proportion
idf_log = np.log(N / df)

# 3. Create the Plot
plt.figure(figsize=(10, 6))

# Plot the blue line (IDF = N/DF)
sns.lineplot(x=df, y=idf_linear, label='IDF = N/DF', linewidth=1.5)

# Plot the orange line (IDF = log(N/DF))
sns.lineplot(x=df, y=idf_log, label='IDF = log(N/DF)', linewidth=1.5)

# 4. Formatting and Labels
plt.title('IDF Values for N equals to 100 documents', pad=15)
plt.xlabel('DF')
plt.ylabel('IDF')

# Set the limits to match the image exactly
plt.ylim(-1, 25) # Slight negative bottom to see the 0 line clearly
plt.xlim(0, 100)

# Set the ticks to match the image (Every 5 units on X-axis)
plt.xticks(np.arange(0, 100, 5))

# Add the legend
plt.legend(loc='upper right', frameon=True)

# Show the final chart
plt.tight_layout()
# Save the updated plot
plt.savefig('idf_log.png', transparent=True)
plt.show()
```

```python
# 2. Generate Data
# The x-axis (Counts of the term) goes from 0 to roughly 3.1 based on the chart
tf_counts = np.linspace(0, 3.1, 500)

# Calculate the y-values for each curve based on the legend formulas
y_raw = tf_counts                       # TF
y_sat_1 = tf_counts / (tf_counts + 1)   # TF / (TF + 1)
y_sat_2 = tf_counts / (tf_counts + 2)   # TF / (TF + 2)
y_sat_3 = tf_counts / (tf_counts + 3)   # TF / (TF + 3)

# 3. Create the Plot
plt.figure(figsize=(10, 6))

# Plotting the lines
# We specify labels manually to match the legend exactly
sns.lineplot(x=tf_counts, y=y_raw, label='TF', linewidth=2)
sns.lineplot(x=tf_counts, y=y_sat_1, label='TF / (TF + 1)', linewidth=2)
sns.lineplot(x=tf_counts, y=y_sat_2, label='TF / (TF + 2)', linewidth=2)
sns.lineplot(x=tf_counts, y=y_sat_3, label='TF / (TF + 3)', linewidth=2)

# 4. Formatting and Labels
plt.title('Term Frequency Values with Saturation terms', pad=15)
plt.xlabel('Counts of the term')
plt.ylabel('Term Value')

# Set ticks to match the image
plt.yticks(np.arange(0, 3.25, 0.25))  # 0.00 to 3.00 steps of 0.25
plt.xticks(np.arange(0, 4, 1))        # 0, 1, 2, 3

# Set limits
plt.ylim(0, 3.0)
plt.xlim(0, 3.2) # Just enough space on the right for the lines to finish

# Legend location
plt.legend(loc='upper left', frameon=True)

# Show the chart
plt.tight_layout()
# Save the updated plot
plt.savefig('tf_saturation.png', transparent=True)
plt.show()
```
