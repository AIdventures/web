```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# IDF Calculations
idf_linear = N / DF
idf_log = np.log(N / DF)

# Plotting with larger fonts and transparent background
plt.figure(figsize=(10, 6))
plt.plot(DF, idf_linear, label='IDF = N/DF', color='cyan')
plt.plot(DF, idf_log, label='IDF = log(N/DF)', color='orange')

# Chart formatting for transparency and larger fonts
plt.title('IDF Values for N equals to 100 documents', color='white', fontsize=16)
plt.xlabel('DF', color='white', fontsize=14)
plt.ylabel('IDF', color='white', fontsize=14)

# Legend with larger font size
plt.legend(facecolor='black', edgecolor='white', labelcolor='white', fontsize=12)

# Grid, ticks, and background adjustments
plt.grid(True, linestyle='--', alpha=0.7, color='white')
plt.gca().spines[:].set_color('white')
plt.gca().tick_params(colors='white', labelsize=12)
plt.gca().set_facecolor('none')  # Transparent background

# Save the updated plot
plt.savefig('idf_values.png', transparent=True)
plt.show()
```
